```python
import numpy as np              # Importing numpy as np
import matplotlib.pyplot as plt # Importing matplotlib.pyplot as plt
```

```python
# Function that takes a, b, c and x values and return the corresponding y values based on relation
# y = ((a / 2) * (x^2)) + (b * x) + c
def getY(a, b, c, x):
    return ((a / 2) * x * x) + (b * x) + c
```

```python
# Q1

# Function to plot function L(w), L* = minVal and w* = minimizer
def plot1(a, b, c, xMin, xMax, minVal, minimizer, partNum):
    x = np.linspace(xMin, xMax, 100)    # Generating x values to plot
    y = getY(a, b, c, x)                # Getting corresponding y values

    # The Label string for the function L(w)
    labelStr = "L(w) = "
    labelStr += ("(" + str(a) + "/2)w^2") if (a != 0) else ""
    labelStr += (" + (" + str(b) + ")w") if (b != 0) else ""
    labelStr += (" + (" + str(c) + ")") if (c != 0) else ""

    plt.plot(x, y, 'r', label = labelStr)   # Plotting function L(w)
    plt.plot(minimizer, minVal, 'bo', label = "L*") # Plotting L* as blue dot
    plt.plot(minimizer, 0, 'g*', label = "w*")      # Plotting w* as green star
    plt.title("Q1 --> (" + str(partNum) + ")")  # Title for the plot
    plt.xlabel("w")     # x-label for the plot
    plt.ylabel("L(w)")  # y-label for the plot
    plt.legend()        # Showing the labels
    plt.draw()          # Drawing the curve
    plt.show()          # Displaying the plot
```

```python
# Q1 -> 1)
a, b, c = 1, 0, 0   # Given values
minVal = 0      # calculated L*
minimizer = 0   # calculated w*
plot1(a, b, c, -5, 5, minVal, minimizer, 1) # Plotting L(w), L*, w* for this case
```

```python
# Q2

# Function to plot complete function L(w) and L(wt) in one plot and wt - w* in other plot
# w* = minimizer, L(wt) = currVal, wt = currW
def plot2(a, b, c, x, y, minimizer, wArr, tArr, partNum):
    plt.plot(x, y, 'r')   # Plotting L(w)

    lwArr = getY(a, b, c, np.array(wArr))   # the L(wt) array for each value of wt
    plt.plot(wArr, lwArr, 'b-o')        # Plotting all the points L(wt)

    plt.title("Q2 --> (" + str(partNum) + ") --> L(w) v/s w")  # Title of the plot
    plt.xlabel("w")     # x-label for the plot
    plt.ylabel("L(w)")  # y-label for the plot
    plt.draw()          # Drawing the curve
    plt.show()          # Displaying the plot

    diffArr = np.array(wArr) - minimizer    # The (wt - w*) array for each value of wt
    plt.plot(tArr, diffArr, 'b-')   # Plotting all the points L(wt)

    plt.title("Q2 --> (" + str(partNum) + ") --> (wt - w*) v/s t")  # Title of the plot
    plt.xlabel("t")     # x-label for the plot
    plt.ylabel("wt - w*")  # y-label for the plot
    plt.draw()          # Drawing the curve
    plt.show()          # Displaying the plot

# Function to perform gradient Descent algorithm for all the parts in Q2
def gradientDescent2(a, b, c, xMin, xMax, minimizer, alpha, T, partNum):
    x = np.linspace(xMin, xMax, 100)    # Generating x values to plot
    y = getY(a, b, c, x)                # Getting corresponding y values

    wArr = []   # Array to store w values
    tArr = []   # iteration number array

    wo = 3  # Taking initial wo
    wArr.append(wo)
    tArr.append(0)

    for t in range(T):  # Performing the algorithm for T iterations
        w = 0   # wt
        if (partNum == 1):  # For part 1
            w = ((1 - alpha) ** (t + 1)) * wo     # Derived relation between wt and wo
        elif (partNum == 2):  # For part 2
            w = ((1 - (0.1 * alpha)) ** (t + 1)) * wo     # Derived relation between wt and wo
        elif (partNum == 3):  # For part 3
            w = ((1 - (10 * alpha)) ** (t + 1)) * wo     # Derived relation between wt and wo
        elif (partNum == 4):  # For part 4
            w = ((1 - alpha) ** (t + 1)) * wo     # Derived relation between wt and wo
        elif (partNum == 5):  # For part 5
            w = (((1 - alpha) ** (t + 1)) * (wo + 1)) - 1     # Derived relation between wt and wo
        elif (partNum == 6):  # For part 6
            w = (((1 - (0.1 * alpha)) ** (t + 1)) * (wo - 10)) + 10     # Derived relation between wt and wo

        wArr.append(w)
        tArr.append(t + 1)
        
    plot2(a, b, c, x, y, minimizer, wArr, tArr, partNum)   # Plotting L(w), wt and wt - w* for each iteration
```

```python
# Q2 -> 1)
# L(w) = (1/2)*w^2
# w(t) = ((1 - alpha)^t)*w(0)
# w* = 0, L* = 0

a, b, c = 1, 0, 0   # Given values
minVal = 0      # calculated L*
minimizer = 0      # calculated w*

# i) Iterates on same side of w* and converges to w*           : 0 < alpha <= 1
# ii) Oscillates on both sides of w* and converges to w*       : 1 < alpha < 2
# iii) Oscillates on both sides of w* and diverges to INFINITY : 2 < alpha

T = 10 # Number of iterations to perform
for alpha in [0.5, 1.5, 3]: # Taking various values of alpha
    print("=================================================")
    gradientDescent2(a, b, c, -5, 5, minimizer, alpha, T, 1)    # Performing gradient descent
    print("=================================================")
```

```python
# Q3
# L(w) = w^4 - w^3 - 5*w^2 + 1  # Taken function

# Function to get the Y values based on the relation
# y = x^4 - x^3 - 5*x^2 + 1
def getY3(x):
    return (x*x*x*x) - (x*x*x) - 5*(x*x) + 1

# Function to get the value of derivative of above function at a given point
def derivative3(w):
    return 4*(w**3) - 3*(w**2) - 10*(w)

x = np.linspace(-5, 5, 100) # Generating x values to plot
y = getY3(x)    # Getting corresponding y values

beta = 0.9  # Beta value for Gradient Descent with Momentum
alpha = 1   # Randomly taken alpha value, initially
wo = -3     # Initial wo

w1 = wo - alpha*(derivative3(wo))   # w1 value for normal gradient descent
# Finding the appropriate value of alpha using Gradient descent with backtrack line search
while (getY3(w1) > getY3(wo)):  # while L(w1) > L(wo)
    alpha *= beta   # Decaying alpha with beta value
    w1 = wo - alpha*(derivative3(wo))   # new value of w1

T = 20 # Number of iterations performed

# Performing normal gradient descent for given function
wt1 = wo    # initial wt for normal gradient descent
for i in range(T):
    wt1 = wt1 - alpha*derivative3(wt1)  # w(t+1) = w(t) - alpha*(derivative(w(t)))

# Performing gradient descent with momentum on given function
wt2 = wo    # initial wt for gradient descent with momentum
v = 0       # velocity
for i in range(T):
    v = (1-beta)*v + beta*(derivative3(wt2))  # v(t) = (1-beta)*v(t-1) - beta*(derivative(w(t)))
    wt2 = wt2 - alpha*v     #   # w(t+1) = w(t) - alpha*v(t)

plt.plot(x, y, 'r-', label = "L(w) = w^4 - w^3 - 5*w^2 + 1")
plt.plot(wt1, getY3(wt1), 'go', label = "L* with normal gradient descent")
plt.plot(wt2, getY3(wt2), 'bo', label = "L* with gradient descent with momentum")
plt.ylim(-12, 8)    # y limits for plot
plt.title("Comparing Gradient Descent and Gradient Descent with Momentum")  # Title of the plot
plt.xlabel("w")     # x-label for the plot
plt.ylabel("L(w)")  # y-label for the plot
plt.legend()    # Displaying the labels
plt.draw()  # Drawing the curve
plt.show()  # Displaying the plot

# Comparing Normal Gradient Descent and Gradient Descent with Momentum, it can be observed that gradient descent with
# momentum depends on the velocity it gains from the slope at the current point. If velocity is higher, it can skip the
# nearest local minima and converges to some other local minima but on the other hand, Normal gradient descent only converges
# to the nearest local minima from the initial starting point. Gradient Descent with Momentum depends upon initial point
# taken as velocities are different at different points and can yield different results.
```

```python
# Q3 - Viva Answer

# Function to plot complete function L(w) and L(wt) in one plot and wt - w* in other plot
# w* = minimizer, L(wt) = currVal, wt = currW
def plot3(x, y, minimizer, wArr, tArr):
    plt.plot(x, y, 'r')   # Plotting L(w)

    lwArr = getY3(np.array(wArr))   # the L(wt) array for each value of wt
    plt.plot(wArr, lwArr, 'b-o')        # Plotting all the points L(wt)

    plt.title("Q3 -->  L(w) v/s w")  # Title of the plot
    plt.xlabel("w")     # x-label for the plot
    plt.ylabel("L(w)")  # y-label for the plot
    plt.draw()          # Drawing the curve
    plt.show()          # Displaying the plot

    diffArr = np.array(wArr) - minimizer    # The (wt - w*) array for each value of wt
    plt.plot(tArr, diffArr, 'b-')   # Plotting all the points L(wt)

    plt.title("Q3 --> (wt - w*) v/s t")  # Title of the plot
    plt.xlabel("t")     # x-label for the plot
    plt.ylabel("wt - w*")  # y-label for the plot
    plt.draw()          # Drawing the curve
    plt.show()          # Displaying the plot

def gradientDescent3(alpha, m1, m2):
    wArr = []   # wArr to store wt values
    tArr = []
    # Performing normal gradient descent for given function
    wt1 = wo    # initial wt for normal gradient descent
    wArr.append(wo)
    tArr.append(0);
    for i in range(T):
        wt1 = wt1 - alpha*derivative3(wt1)  # w(t+1) = w(t) - alpha*(derivative(w(t)))
        wArr.append(wt1)
        tArr.append(i + 1)

    minimizer = m1
    print("========== Normal Gradient Descent ==========")
    plot3(x, y, minimizer, wArr, tArr)   # Plotting L(w), wt and wt - w* for each iteration

    wArr = []   # wArr to store wt values
    tArr = []
    # Performing gradient descent with momentum on given function
    wt2 = wo    # initial wt for gradient descent with momentum
    v = 0       # velocity
    wArr.append(wo)
    tArr.append(0);
    for i in range(T):
        v = (1-beta)*v + beta*(derivative3(wt2))  # v(t) = (1-beta)*v(t-1) - beta*(derivative(w(t)))
        wt2 = wt2 - alpha*v     #   # w(t+1) = w(t) - alpha*v(t)
        wArr.append(wt2)
        tArr.append(i + 1)
            
    minimizer = m2
    print("========== Gradient Descent with Momentum ==========")
    plot3(x, y, minimizer, wArr, tArr)   # Plotting L(w), wt and wt - w* for each iteration

print("================= No Oscillation =================")
gradientDescent3(0.01, -1.25, -1.25)
print("================= Oscillation and Converging =================")
gradientDescent3(0.05814973700304011, 2, 2)
# print("================= Oscillation and Diverging =================")
# gradientDescent3(0.6, -1.25, 2) # The numerical result was going out of range to plot the curve
```

```python
# Q4

# Representing 8 direction points in w(1) and w(2) axis
dw1 = [4, 4, 0, -4, -4, -4, 0, 4]
dw2 = [0, 4, 4, 4, 0, -4, -4, -4]

# Function to get the partial derivative at a given w for the parts in 4th question
# dim == 0, partial derivative w.r.t. w(1)
# dim == 1, partila derivative w.r.t. w(2)
def derivative4(w, dim, partNum):
    if (partNum == 1):  # f1
        return w

    elif (partNum == 2):  # f2
        return (10 * w) if (dim == 0) else (w)

    elif (partNum == 3):  # f3
        return (w) if (dim == 0) else (10 * w)

    elif (partNum == 4):  # f4
        return (w + 5) if (dim == 0) else (w - 3)

# Function to perform gradient descent algorithm for various parts of Q4
def gradientDescent4(initialWo, alpha, T, partNum):
    wo = [*initialWo]   # Taking initial wo
    wArr = []           # To store w Values
    wArr.append([*wo])  # Storing wo
    wt = [*wo]  # wt 
    for t in range(T):  # Executing for T iterations
        wt[0] = wt[0] - alpha*derivative4(wt[0], 0, partNum)    # w[0], w.r.t. w(1)
        wt[1] = wt[1] - alpha*derivative4(wt[1], 1, partNum)    # w[1], w.r.t. w(2)
        wArr.append([*wt])  # Storing w

    plt.plot(np.array(wArr)[:, 0], np.array(wArr)[:, 1], 'r-o') # Plotting w values obtained
    plt.draw()  # Drawing the curve
    plt.show()  # Displaying the plot
    
```
