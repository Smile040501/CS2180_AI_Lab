```python
# Q1

# Function that take two points and returns the Euclidean Distance between those two points
def findDist(point1, point2):
    p1 = np.array([point1[0], point1[1]])   # Taking the x and y coordinates of point1
    p2 = np.array([point2[0], point2[1]])   # Taking the x and y coordinates of point2

    return np.linalg.norm(p1-p2)


# Function to find k nearest neighbours in the given dataset from a given point
def findKNearestNeighbours(dataset, point, k):
    numPoints = len(dataset) # number of data points

    if numPoints <= k:    # if number of points are less than or equal to k, return them as answer
        return dataset

    distanceArr = []    # Array to store the [index, distance]
    # where, index = index of the data point and, distance = distance of the data point from the given point
    for i in range(numPoints):  # Storing distance of all the data points from the given point
        distanceArr.append([i, findDist(dataset[i], point)])
        
    distanceArr.sort(key = lambda p: p[1])   # Sorting the distanceArr on the basis of distance of points

    kNearestPoints = [] # Array to store only first k points from the distanceArr
    for i in range(k):  # Taking first k points as the knn and storing them
        index = distanceArr[i][0]
        kNearestPoints.append([*dataset[index]])
        
    return kNearestPoints
```

```python
# Function to predict the result from the knn found based on majority
def getPrediction(kNearestNeighbours):
    count1 = 0  # Count of data points with -1 as label
    count2 = 0  # Count of data points with 1 as label

    for point in kNearestNeighbours:    # Looping all the knn
        if (point[2] == -1):    # label is -1
            count1 += 1
        else:                   # label is 1
            count2 += 1

    if(count1 != count2):       # if count of both are not equal
        return -1 if (count1 > count2) else 1   # Return the one with the majority

    return np.random.choice([1, -1], p = [0.5, 0.5])    # Return any random from them with equal probability of both

# Function to measure the accuracy of the KNN on the test data and returns % of correct predictions made
def knnAcc(k):
    correct = 0     # Count of number of correct predictions

    for point in knnTest:   # Looping all the points in the knn-test data
        kNearestNeighbours = findKNearestNeighbours(knnTrain, point, k) # Finding knn for that point
        predictedValue = getPrediction(kNearestNeighbours)  # Getting the predicted value for that point
        if (predictedValue == point[2]):        # if predicted value is equal to actual value, it is correct prediction
            correct += 1
    
    return (correct / len(knnTest)) * 100      # returning the percentage of correct predictions made
```

```python
# Q2

# Function to pad the given dataset and w vector by 1
# Each point in dataset is of form [x1, x2, ... , xn, y]
# w vector is of form [w1, w2, ... wn]
def padData(dataset, w):
    newDataset = []
    for point in dataset:
        newPoint = []
        for i in range(len(point) - 1): # Adding first len - 1 values of point, basically [x1, x2, ... , xn]
            newPoint.append(point[i])

        newPoint.append(1)              # Padding by 1
        newPoint.append(point[len(point) - 1])  # Adding y to the point
        newDataset.append(newPoint) # Adding the padded point to the newDataset
    
    newW = []
    for i in range(len(w)): # Adding all the values of w
        newW.append(w[i])
        
    newW.append(0)          # Padding it by 0

    return [newDataset, newW]   # Returning the final newDataset and newW vector
```

```python
# Function to calculate the value of R by which the data points are assumed to be bounded
def calculateR(dataset):
    maxVal = 0
    for point in dataset:   # Finding the maximum value of norm(point)
        norm = np.linalg.norm(point[: -1])
        if (maxVal < norm): # Updating maxVal
            maxVal = norm

    return maxVal
```

```python
# Function to calculate the value of Gamma, which is minimum( yi * (w.xi) / norm(w))
def calculateGamma(dataset, w):
    norm = np.linalg.norm(w)    # norm of w vector
    firstPoint = dataset[0]     # First data point
    minVal = firstPoint[len(firstPoint) - 1] * (np.dot(w, firstPoint[: -1]) / norm) # Initial value to start with

    for point in dataset:       # Iterating all the points
        val = point[len(point) - 1] * (np.dot(w, point[: -1]) / norm)   # Finding their value
        if (val < minVal):      # Updating minVal
            minVal = val
            
    return minVal
```

```python
# Function to check if the given vector correctly fits the data points or not 
# i.e. the algoithm converge or not i.e. there are no misclassified points left
def doConverge(dataset, w):
    for point in dataset:
        x = point[: -1]     # xi value
        y = point[len(point) -1]    # yi value
        if (np.dot(w, x) * y <= 0): # if yi * (w.xi) <= 0, then point is misclassified
            return [False, [*point]]

    return [True, []]
```

```python
# Function to perform the perceptron algorithm, given dataset and if it is biased or not
def perceptron(dataset, biased = False):
    w = list(np.zeros(len(dataset[0]) - 1))     # Initial w vector
    R = calculateR(dataset) # Calculating R value of the dataset

    if (biased):    # if dataset is biased, pad it by 1 technique
        [dataset, w] = padData(dataset, w)
        
    iterations = 0  # Number of iterations taken
    [converged, misclassified] = doConverge(dataset, w)  # Checking if there is any misclassified point

    while (not converged):  # while there exists a misclassified point
        x = np.array(misclassified[: -1])   # xi of misclassified point
        y = misclassified[len(misclassified) - 1]   # yi of misclassified point
        w = list(np.array(w) + (y*x))   # updating w as w = w + yi*xi

        iterations += 1     # Incrementing iterations
        [converged, misclassified] = doConverge(dataset, w) # Checking if there is any misclassified point

    Gamma = calculateGamma(dataset, w) # Calculating Gamma for the perceptron algorithm
    return [w, R, Gamma, iterations]
```
